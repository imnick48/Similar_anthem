```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
preprocessed_data=pd.read_csv('preprocessed_dataset.csv')
preprocessed_data.head()
```

```python
kmeans=KMeans(n_clusters=5)
clusters=kmeans.fit_predict(preprocessed_data.drop('Country',axis=1))+1
```

```python
out_df=pd.DataFrame()
out_df['Country']=preprocessed_data['Country']
out_df['Clusters']=clusters
out_df.to_csv('Clusters.csv',index=False)
```
