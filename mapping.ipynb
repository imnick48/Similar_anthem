```python
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
```

```python
cluster=pd.read_csv('Clusters.csv')
cluster.head()
```

```python
#If you are using windows use \\ instead of //.
world = gpd.read_file("110m_cultural//ne_110m_admin_0_countries.shp")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world = world.merge(cluster, how="left", left_on="NAME", right_on="Country")
world.boundary.plot(ax=ax)
world.plot(column="Clusters",ax=ax,legend=True)
plt.savefig("clustered_map.png", dpi=300, bbox_inches='tight')
plt.show()
```
